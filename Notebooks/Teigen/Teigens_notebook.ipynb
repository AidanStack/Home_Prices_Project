```python
#Determining rough estimate of number of properties/records which will be considered multi 
#or 2+ story units. Also using the value counts to determine how to group the houses.
#Grouping will be done in two ways, the first, by whether the house is a multi-story
#unit (more than 1 floor) and the second, by whether the property is a true two story or
#more property.

df['floors'].value_counts()
```

```python
#For loop to create a list of values indicating whether each house was a single or multi
#story unit.

result = []
for i in df['floors']:
    if i > 1.0:
        result.append(1)
    else:
        result.append(0)
```

```python
#Creating new column in df indicating multi-story or not.

df['multi'] = result
```

```python
#Checking to ensure that the new column was properly inserted into the df.

df
```

```python
#For loop that shows whether the unit in question was 2+ stories or not.
#A value of 0 indicates that the unit was less than 2 stories and a value of 1 indicates
#that the unit was 2 or more stories.

result2 = []
for i in df['floors']:
    if i >= 2.0:
        result2.append(1)
    else:
        result2.append(0)
```

```python
#Creating a new column in df indicating whether the unit is a tru multi story (greater
#than or equal to 2 floors) or not by assigning the results from the most previous
#for loop as values.

df['two_or_more_story'] = result2
```

```python
#Esnuring the 'two_or_more_story' column was added properly.

df
```

```python
#Calling value counts method on the 'sqft_basement' column of the df to determine a 
#rough estimate as to the number of houses which do not have a basement. From looking 
#at the value counts, it is clear that we should have at least 13,280 properties or 
#records which do not have a basement.

df['sqft_basement'].value_counts()
```

```python
#For loop to indicate whether each house had a basement or not. A value of 0 indicates
#that the house does not have a basement, whereas a value of 1 indicates the presence
#of a basement.

result3 = []
for i in df['sqft_basement']:
    if i == '0.0':
        result3.append(0)
    elif i == '?':
        result3.append(0)
    else:
        result3.append(1)
```

```python
#Assigning the results from the most previous for loop to a new column in df named 
#'basement' which will be used to determine whether each house has a basement or not.

df['basement'] = result3
```

```python
#Ensuring the proper number of each value by cross referencing with the value counts
#method called on the 'sqft_basement' column done above. Seems as though everything checks
#out, as the number of '0.0' or '?' values from the value counts cell is 13,280, which
#is the exact number of '0' values in our new 'basement' column.

df['basement'].value_counts()
```
